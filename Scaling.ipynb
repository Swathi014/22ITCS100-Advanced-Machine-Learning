```python
import numpy as np
import random
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
y = np.random.random(300)
random.seed(300)
x = np.arange(300)
print(x.shape)
print(y.shape)
```

#Standardization with python

```python
def Standardization(X):
  mean = np.mean(X,axis = 0)
  std = np.std(X,axis = 0)
  X_std = (X-mean)/std
  return X_std
```

```python
Y = Standardization(y)
```

```python
fig, (ax1,ax2) = plt.subplots(ncols = 2, figsize = (6,5))
ax1.set_title('Before Standardization')
sns.kdeplot(y,ax = ax1)
ax2.set_title('After Standardization')
sns.kdeplot(Y,ax = ax2)
plt.show()
```

#Standardization with sklearn

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
y = y.reshape(-1,1)
Y_std = scalar.fit_transform(y)
Y_std = Y_std.flatten()
```

```python
fig, (ax1,ax2) = plt.subplots(ncols = 2, figsize = (10,5))
ax1.set_title('After manual Standardization')
sns.kdeplot(Y,ax = ax1)
ax2.set_title('After scikit Standardization')
sns.kdeplot(Y_std,ax = ax2)
plt.show()
```

#Normalization with python

```python
def normalize(X):
  Xmin=np.max(X)
  Xmax=np.min(X)

  X_norm=(X-Xmin)/(Xmax-Xmin)
  return X_norm
```

```python
Y_norm = normalize(y)
Y_norm = Y_norm.flatten()
```

#Normalization with scikit_learn

```python
from sklearn.preprocessing import MinMaxScaler
y = y.reshape(-1,1)
scalar = MinMaxScaler()
Y_minmax = scalar.fit_transform(y)
Y_minmax = Y_minmax.flatten()
print(Y_norm.shape,Y_minmax.shape)
```

```python
fig, (ax1,ax2) = plt.subplots(ncols = 2, figsize = (10,5))
ax1.set_title('After manual Normalization')
sns.kdeplot(Y_norm,ax = ax1)
ax2.set_title('After scikit Normalization')
sns.kdeplot(Y_minmax,ax = ax2)
plt.show()
```
