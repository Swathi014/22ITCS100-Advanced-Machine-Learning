```python
import matplotlib.pyplot as plt
import numpy as np
import os
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
data = np.loadtxt("/content/drive/MyDrive/MTECH/Dataset/ex1data1.csv", delimiter=",")
X,y = data[:,0],data[:,1]
m = y.size
```

```python
print("Plotting data ...\n")
def plotData(x, y):
    plt.figure(figsize=(7,5))
    plt.plot(x, y, 'ro', ms=8, mec='k')
    plt.xlabel('Population (10,000s)')
    plt.ylabel('Profit ($10,000)')
    plt.grid(True)
    plt.show()
```

```python
plotData(X,y)
```

```python
print("the first 5 samples {}".format(X[:5]))
print("The first five targets {}".format(y[:5]))
X = X.reshape(-1,1)
y = y.reshape(-1,1)
print("The no. of samples in train set is {}".format(X.shape))
print("The no. of samples in test set is {}".format(y.shape))
lr = LinearRegression()
lr.fit(X,y)
```

# Prediction and Model Evaluation

```python
y_pred = lr.predict(X)
plt.scatter(X,y,color = "b")
plt.plot(X,y_pred,color = "k")
```
